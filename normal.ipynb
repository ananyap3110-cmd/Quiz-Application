```python
quiz = {
    "What is the output of print(type([]))?": {
        "options": ["A) <class 'tuple'>", "B) <class 'list'>", "C) <class 'dict'>", "D) <class 'set'>"],
        "answer": "B"
    },
    "Which keyword is used to define a function in Python?": {
        "options": ["A) func", "B) define", "C) def", "D) function"],
        "answer": "C"
    },
    "What does len('hello') return?": {
        "options": ["A) 4", "B) 6", "C) 5", "D) None"],
        "answer": "C"
    },
    "Which of these is a mutable data type?": {
        "options": ["A) tuple", "B) string", "C) int", "D) list"],
        "answer": "D"
    },
    "What is the result of 10 // 3?": {
        "options": ["A) 3.33", "B) 3", "C) 4", "D) 1"],
        "answer": "B"
    },
    "Which method adds an element to the end of a list?": {
        "options": ["A) add()", "B) insert()", "C) append()", "D) push()"],
        "answer": "C"
    },
    "What symbol is used for single-line comments in Python?": {
        "options": ["A) //", "B) /* */", "C) --", "D) #"],
        "answer": "D"
    },
    "What is the output of bool(0)?": {
        "options": ["A) True", "B) False", "C) None", "D) Error"],
        "answer": "B"
    },
    "Which of these is used to handle exceptions in Python?": {
        "options": ["A) try/except", "B) catch/throw", "C) error/handle", "D) check/fix"],
        "answer": "A"
    },
    "What does the 'in' keyword check?": {
        "options": ["A) Variable type", "B) Membership in a sequence", "C) Integer range", "D) Indentation"],
        "answer": "B"
    }
}


def run_quiz():
    print("=" * 50)
    print("       🐍 PYTHON QUIZ APPLICATION 🐍")
    print("=" * 50)
    print(f"Total Questions: {len(quiz)}")
    print("Choose the correct option (A / B / C / D)")
    print("Type 'EXIT' at any time to quit the quiz.")
    print("=" * 50)

    score = 0
    wrong_answers = []

    for q_num, (question, data) in enumerate(quiz.items(), start=1):
        print(f"\nQ{q_num}: {question}")
        for option in data["options"]:
            print(f"   {option}")

        while True:
            user_input = input("Your answer (or type EXIT to quit): ").strip().upper()
            if user_input == "EXIT":
                print("\n[!] You chose to exit the quiz.")
                if score > 0 or q_num > 1:
                    print(f"  Questions Attempted : {q_num - 1}")
                    print(f"  Score so far        : {score} / {len(quiz)}")
                print("Thanks for playing! See you next time.")
                print("=" * 50)
                return
            if user_input in ["A", "B", "C", "D"]:
                break
            print("  [!] Invalid input! Please enter A, B, C, or D (or EXIT to quit).")

        if user_input == data["answer"]:
            print("✅ Correct!")
            score += 1
        else:
            correct_option = data["answer"]
            correct_text = next(opt for opt in data["options"] if opt.startswith(correct_option))
            print(f"❌ Wrong! Correct answer: {correct_text}")
            wrong_answers.append((f"Q{q_num}: {question}", correct_text))

    # Results
    print("\n" + "=" * 50)
    print("               📊 QUIZ RESULTS")
    print("=" * 50)
    print(f"  Score      : {score} / {len(quiz)}")
    percentage = (score / len(quiz)) * 100
    print(f"  Percentage : {percentage:.1f}%")

    if percentage == 100:
        grade = "🏆 Outstanding!"
    elif percentage >= 80:
        grade = "🥇 Excellent!"
    elif percentage >= 60:
        grade = "🥈 Good Job!"
    elif percentage >= 40:
        grade = "🥉 Keep Practicing!"
    else:
        grade = "📚 Needs Improvement"

    print(f"  Grade      : {grade}")

    if wrong_answers:
        print("\n--- Review Wrong Answers ---")
        for q, ans in wrong_answers:
            print(f"  {q}")
            print(f"     ➤ Correct: {ans}")

    print("=" * 50)
    print("Thanks for playing! Keep learning Python! 🐍")
    print("=" * 50)


if __name__ == "__main__":
    run_quiz()
```
