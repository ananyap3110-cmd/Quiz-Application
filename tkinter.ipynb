```python
import tkinter as tk
import random

# Quiz Data
quiz = {
    "easy": {
        "What is the capital of France?": {
            "options": ["Berlin", "Madrid", "Paris", "Rome"],
            "answer": "Paris"
        },
        "Which planet is Red Planet?": {
            "options": ["Earth", "Mars", "Jupiter", "Venus"],
            "answer": "Mars"
        }
    },
    "medium": {
        "Which language is used for web apps?": {
            "options": ["Python", "JavaScript", "C++", "Java"],
            "answer": "JavaScript"
        },
        "What does CPU stand for?": {
            "options": ["Central Process Unit", "Central Processing Unit", "Computer Unit", "Processor Unit"],
            "answer": "Central Processing Unit"
        }
    },
    "hard": {
        "Who developed Python?": {
            "options": ["Elon Musk", "Bill Gates", "Guido van Rossum", "Mark Zuckerberg"],
            "answer": "Guido van Rossum"
        },
        "Binary search complexity?": {
            "options": ["O(n)", "O(log n)", "O(n log n)", "O(1)"],
            "answer": "O(log n)"
        }
    }
}


class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Quiz App")
        self.root.geometry("500x400")

        self.score = 0
        self.q_index = 0
        self.questions = []

        self.create_start_screen()

    def clear_screen(self):
        for widget in self.root.winfo_children():
            widget.destroy()

    def create_start_screen(self):
        self.clear_screen()

        tk.Label(self.root, text="Select Difficulty", font=("Arial", 16)).pack(pady=20)

        tk.Button(self.root, text="Easy", width=15, command=lambda: self.start_quiz("easy")).pack(pady=5)
        tk.Button(self.root, text="Medium", width=15, command=lambda: self.start_quiz("medium")).pack(pady=5)
        tk.Button(self.root, text="Hard", width=15, command=lambda: self.start_quiz("hard")).pack(pady=5)

    def start_quiz(self, level):
        self.score = 0
        self.q_index = 0
        self.questions = list(quiz[level].items())
        random.shuffle(self.questions)

        self.show_question()

    def show_question(self):
        self.clear_screen()

        if self.q_index >= len(self.questions):
            self.show_result()
            return

        question, data = self.questions[self.q_index]

        tk.Label(self.root, text=question, font=("Arial", 14), wraplength=400).pack(pady=20)

        self.selected = tk.StringVar()

        for option in data["options"]:
            tk.Radiobutton(self.root, text=option, variable=self.selected, value=option).pack(anchor="w")

        tk.Button(self.root, text="Next", command=self.check_answer).pack(pady=20)

    def check_answer(self):
        question, data = self.questions[self.q_index]

        if self.selected.get() == data["answer"]:
            self.score += 1

        self.q_index += 1
        self.show_question()

    def show_result(self):
        self.clear_screen()

        tk.Label(self.root, text="Quiz Finished!", font=("Arial", 16)).pack(pady=20)
        tk.Label(self.root, text=f"Score: {self.score}/{len(self.questions)}", font=("Arial", 14)).pack(pady=10)

        tk.Button(self.root, text="Play Again", command=self.create_start_screen).pack(pady=10)
        tk.Button(self.root, text="Exit", command=self.root.quit).pack(pady=5)


# Run App
root = tk.Tk()
app = QuizApp(root)
root.mainloop()
```
